```python
df = pd.read_csv("../data/processed/EMS_Cleaned.csv")

datetime_cols = [
    "incident_datetime",
    "first_assignment_datetime",
    "first_activation_datetime",
    "first_on_scene_datetime",
    "first_to_hosp_datetime",
    "incident_close_datetime"
]

print(df.shape)
df.head()
```

```python
df["peak_hour"] = (
    ((df["hour"] >= 7) & (df["hour"] <= 10)) |
    ((df["hour"] >= 16) & (df["hour"] <= 19))
).astype(int)
```

```python
df["night_shift"] = (
    (df["hour"] >= 22) |
    (df["hour"] <= 5)
).astype(int)
```

```python
print(df[["first_activation_datetime", "first_on_scene_datetime"]].dtypes)
```

```python
print(df["first_activation_datetime"].head(10))
print(df["first_on_scene_datetime"].head(10))
```

```python
datetime_cols = [
    "incident_datetime",
    "first_assignment_datetime",
    "first_activation_datetime",
    "first_on_scene_datetime",
    "first_to_hosp_datetime",
    "incident_close_datetime"
]

for col in datetime_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")
```

```python
print(df["first_activation_datetime"].head())
print(df["first_on_scene_datetime"].head())
```

```python
df["travel_duration"] = (
    df["first_on_scene_datetime"] -
    df["first_activation_datetime"]
).dt.total_seconds()

print(df["travel_duration"].head())
```

```python
df["hospital_time"] = (
    df["first_to_hosp_datetime"] -
    df["first_on_scene_datetime"]
).dt.total_seconds()
```

```python
df["closure_time"] = (
    df["incident_close_datetime"] -
    df["first_to_hosp_datetime"]
).dt.total_seconds()
```

```python
df["zipcode_frequency"] = df["zipcode"].map(
    df["zipcode"].value_counts()
)
```

```python
df["borough_frequency"] = df["borough"].map(
    df["borough"].value_counts()
)
```

```python
df["dispatch_area_frequency"] = df["incident_dispatch_area"].map(
    df["incident_dispatch_area"].value_counts()
)
```

```python
df["initial_call_frequency"] = df["initial_call_type"].map(
    df["initial_call_type"].value_counts()
)
```

```python
df["final_call_frequency"] = df["final_call_type"].map(
    df["final_call_type"].value_counts()
)
```

```python
df["rush_hour"] = df["hour"].isin([7, 8, 9, 16, 17, 18, 19]).astype(int)

df["office_hours"] = df["hour"].between(9, 17).astype(int)

df["call_borough"] = (
    df["initial_call_type"].astype(str) + "_" +
    df["borough"].astype(str)
).astype("category").cat.codes

df["zip_hour"] = (
    df["zipcode"].astype(str) + "_" +
    df["hour"].astype(str)
).astype("category").cat.codes

print("New Features Added Successfully!")
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:

    df[col] = df[col].fillna(df[col].median())
```

```python
df.drop(columns=datetime_cols, inplace=True, errors="ignore")
```

```python
import joblib
import os

os.makedirs("artifacts", exist_ok=True)

zipcode_frequency = df["zipcode"].value_counts().to_dict()
borough_frequency = df["borough"].value_counts().to_dict()
initial_call_frequency = df["initial_call_type"].value_counts().to_dict()
final_call_frequency = df["final_call_type"].value_counts().to_dict()
dispatch_area_frequency = df["incident_dispatch_area"].value_counts().to_dict()

joblib.dump(zipcode_frequency, "artifacts/zipcode_frequency.pkl")
joblib.dump(borough_frequency, "artifacts/borough_frequency.pkl")
joblib.dump(initial_call_frequency, "artifacts/initial_call_frequency.pkl")
joblib.dump(final_call_frequency, "artifacts/final_call_frequency.pkl")
joblib.dump(dispatch_area_frequency, "artifacts/dispatch_area_frequency.pkl")
```

```python
call_borough_mapping = {
    k: i for i, k in enumerate(
        (df["initial_call_type"].astype(str) + "_" +
         df["borough"].astype(str)).astype("category").cat.categories
    )
}

joblib.dump(
    call_borough_mapping,
    "artifacts/call_borough_mapping.pkl"
)
```

```python
zip_hour_mapping = {
    k: i for i, k in enumerate(
        (df["zipcode"].astype(str) + "_" +
         df["hour"].astype(str)).astype("category").cat.categories
    )
}

joblib.dump(
    zip_hour_mapping,
    "artifacts/zip_hour_mapping.pkl"
)
```
